```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.listdir('../store')
```

```python
with_grid = np.load('../store/fr_4p4pitch_3.8pix_nogrid_10pathsperpixel.npy')
without_grid = np.load('../WITH_WITHOUT_GRID/store_nogrid/fr_4p4pitch_3.8pix_nogrid_10pathsperpixel.npy')
# without_grid = np.load('response_5x5_no_shield.npz')['response']
# with_grid_jiangmei = np.load('response_5x5_shield_1000V.npz')['response']
```

```python
with_grid.shape, without_grid.shape
```

```python
with_grid = with_grid.reshape((25,25,4199))*1000
without_grid = without_grid.reshape((25,25,6399))*1000
```

```python
# shift = np.argmax(with_grid[0,0]) - np.argmax(without_grid[0,0])
shift = np.argmax(with_grid[0,0]) - np.argmax(without_grid[0,0])
```

```python
def plot_fr(with_grid, without_grid, N=5, k=0, shift=0):
    plt.subplots(N-k,N-k, figsize=(20,20))
    for i in range(k, N):
        for j in range(k, N):
            plt.subplot(N-k,N-k,(i-k)*(N-k)+j-k+1)
            # plt.plot(np.arange(fr_old.shape[2]), fr_old[i,j], label=f'Before change without grid at [{i},{j}]')
            plt.plot(np.arange(with_grid.shape[2])-shift, with_grid[i,j], label=f'with grid at [{i},{j}]')
            plt.plot(np.arange(without_grid.shape[2]), without_grid[i,j],'--',label=f'without grid result, at [{i},{j}]')
            # plt.plot(np.arange(fr_new.shape[2]), fr_new[i,j], label=f'After change without grid at [{i},{j}]')
            # plt.plot(np.arange(pochoir_result.shape[2])-shift, pochoir_result[i,j],'--', label=f'with grid result, at [{i},{j}]')
            # plt.xlim(1500, 1900)
            # plt.xlim(3500, 4000)
            # plt.xlim(2000, 4000)
            plt.xlim(1500, 2000)
            plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_fr(with_grid, without_grid, N=5, k=0, shift=shift)
```

```python
plot_fr(with_grid, without_grid, N=10, k=5, shift=shift)
```

```python
plot_fr(with_grid, without_grid, N=15, k=10, shift=shift)
```

```python
plot_fr(with_grid, without_grid, N=20, k=15, shift=shift)
```
